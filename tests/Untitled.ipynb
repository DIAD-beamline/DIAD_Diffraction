```python

from PyPDF2 import PdfMerger

# Create a merger object
merger = PdfMerger()

# List the files in the order you want them merged
pdf_files = ["C:/Users/cxh52226/OneDrive - Diamond Light Source Ltd/DIAD\Papers\OWN\Allegra\Nanomaterials/4-mdpi-disclosure-form ML-1.pdf",
             "C:/Users/cxh52226/OneDrive - Diamond Light Source Ltd\DIAD\Papers\OWN\Allegra\Nanomaterials/4-mdpi-disclosure-form ML-2.pdf"]

# Append each PDF to the merger
for pdf in pdf_files:
    merger.append(pdf)

# Write the output to a new PDF
merger.write("C:\Users\cxh52226\OneDrive - Diamond Light Source Ltd\DIAD\Papers\OWN\Allegra\Nanomaterials/combined.pdf")
merger.close()

print("PDFs successfully merged into combined.pdf")
```
